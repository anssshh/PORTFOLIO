# Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
from skimage import io
import datetime as dt
import plotly.express as px
import seaborn as sns
```

# Data Pre-Processing

```python
# Read CSV

data = pd.read_csv('/raw_air_pollution_data_2023.csv')
data
```

```python
# Read Data Description

#dd = pd.read_csv('Metadata.csv')
#dd
```

```python
data['date'] = pd.to_datetime(data['date'])
print(data)
```

```python
data['Tanggal'] = data['date'].dt.strftime('%Y-%m-%d')
data['Waktu'] =data['date'].dt.strftime('%H')
data['Tanggal'] = pd.to_datetime(data['Tanggal'])

print(data)
```

```python
data = data.drop('date', axis=1)
```

```python
data.isna().sum()
```

```python
daftar_lokasi = data['location'].unique()

for lokasi in daftar_lokasi:
    print(lokasi)
```

```python
# Memisahkan data berdasarkan lokasi dan menyimpannya dalam file terpisah
for lokasi in daftar_lokasi:
    data_lokasi = data[data['location'] == lokasi]
    nama_file = f'data_{lokasi}.csv'  # Nama file baru berdasarkan lokasi
    data_lokasi.to_csv(nama_file, index=False)
```

```python
# Data Ancol

data_Ancol = pd.read_csv('data_Ancol.csv')
data_Ancol
```

```python
data_Ancol.fillna(data_Ancol['temperature'].median(), inplace = True)
data_Ancol.fillna(data_Ancol['humidity'].median(), inplace = True)
data_Ancol.fillna(data_Ancol['um050'].median(), inplace = True)
data_Ancol.fillna(data_Ancol['um003'].median(), inplace = True)
data_Ancol.fillna(data_Ancol['um005'].median(), inplace = True)
data_Ancol.fillna(data_Ancol['pressure'].median(), inplace = True)
data_Ancol
```

Histogram Rank pm1 & pm25

```python
# Menghitung jumlah pm1 berdasarkan lokasi
df = merged_data.groupby('location')['pm1'].mean().reset_index()
df = df.sort_values(by='pm1', ascending=False)
df1 = merged_data.groupby('location')['pm25'].mean().reset_index()
df1 = df1.sort_values(by='pm25', ascending=False)
```

Histogram Rank pm1

```python
#Histogram pm1
df_histogram = df.groupby('location', as_index=False)[['pm1']].mean()
df_histogram = df_histogram.sort_values(by='pm1', ascending=True)
fig = px.histogram(df_histogram, x='location', y='pm1',
                   color='location',
                   color_discrete_sequence=px.colors.sequential.Greens_r)

fig.update_layout(height=500, width=700, showlegend=True)
fig.update_xaxes(title_text='Rank Wilayah berdasarkan jumlah pm1')

fig.show()
```

```python
# Membuat histogram
df1_histogram = df1.groupby('location', as_index=False)[['pm25']].mean()
df1_histogram = df1_histogram.sort_values(by='pm25', ascending=True)
fig = px.histogram(df1_histogram, x='location', y='pm25',
                   color='location',
                   color_discrete_sequence=px.colors.sequential.Blues_r)

fig.update_layout(height=500, width=700, showlegend=True)
fig.update_xaxes(title_text='Rank Wilayah berdasarkan jumlah pm25')

fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
lim_cat = [0, 15.5, 55.4, 150.4, 250.4, np.inf]
cat = ['Baik', 'Sedang', 'Tidak Sehat', 'Sangat Tidak Sehat', 'Berbahaya' ]
df1_histogram['category'] = pd.cut(df1_histogram['pm25'], bins=lim_cat, labels=cat)
plt.figure(figsize=(10,6))
colors = plt.cm.get_cmap('viridis', len(cat))
bars = plt.bar(df1_histogram['location'], df1_histogram['pm25'], color=colors(np.arange(len(df1_histogram))))
plt.xlabel('lokasi')
plt.ylabel('mean pm25')
plt.title('Kategori Polusi Udara Masing-masing Daerah')

for bar, kategori in zip(bars, df1_histogram['category']):
  plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 5, kategori, ha='center')

plt.show()
```
